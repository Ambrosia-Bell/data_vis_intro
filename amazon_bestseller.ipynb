## The Process

To start the process I will first import the packages needed to bring this analysis to life.

```python
import matplotlib.pyplot as plt
import csv
import datetime
```

Now I will bring in the CSV data and parse it. Then I will have the python program read the file and close it.

```python
with open('bestsellers with categories.csv', 'r', encoding='utf-8') as datafile:
    amazon_book_data = csv.DictReader(datafile)
    book_data = []
    
    for row in amazon_book_data:
        book_data.append(row)
```
